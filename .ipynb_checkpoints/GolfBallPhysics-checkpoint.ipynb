## Analysis Tools

```python
def showData(fn):
    rate, data = read(fn)
    data = data[:, np.newaxis]

    rms = np.sqrt(np.sum(data**2) / data.shape[0])
    rms = np.full(data.shape, rms)

    print('Sample Rate:', rate)
    plt.plot(data)
    plt.plot(2*rms)
    plt.plot(-2*rms)
    plt.show()
    
    return data, rate
    
fn = 'data\\top_flite_1_4.wav'
data, rate = showData(fn)
```

```python
def meanReduce(data, n=1):
    m = data.shape[0]
    dat = np.copy(data)

    for i in range(n):
        rms = np.sqrt(np.sum(dat**2) / m)

        dat = dat * (1 - rms / np.abs(dat + 1e-16))
    
    return dat

dat = meanReduce(data, n=5)
plt.plot(np.abs(dat))
plt.show()
```

```python
def absMeanReduce(data, n=100):
    m = data.shape[0]
    dat = np.abs(data)

    for i in range(n):
        rm = np.sum(np.abs(dat)) / m
    #     print('Iteration:', i+1, rm)

        dat = dat * (1 - rm / np.abs(dat + 1e-16))

    return dat
    
dat = absMeanReduce(data, 300)
plt.plot(dat)
plt.show()
```

```python
def rmsCutoff(data, k=5):
    m = data.shape[0]
    dat = np.abs(data)

    mean = np.sum(np.abs(dat)) / m
    dat = np.where(dat < k*mean, 0, dat)
    
    return dat

dat = rmsCutoff(data, k=6)
plt.plot(dat)
plt.show()
```

```python
def trimSignal(signal):
    ti = signal[0]
    started = False
    left = 0
    for i in range(len(signal)):
        tf = signal[i]

        # Set left trim once signal has started decreasing.
        if tf < ti and started == False:
            started = True
            left = i - 1
            
        # Set right trim if signal starts increasing.
        if tf > ti and started == True:
            return signal[left:i-1]
        
        ti = tf
        
    return signal[left:]
```

```python
def verifyFit(data, rate):
    # time for object to freefall 1 cm on earth
    h = 0.01
    g = 9.80665
    min_dt = np.sqrt(8 * h / g) * rate
    print('Min Frame Distance:', min_dt)

    dat = rmsCutoff(data, k=6)
    x = dat.flatten()
    peaks, _ = find_peaks(x, distance=min_dt)
    deriv = np.diff(peaks) / rate
    trimmed_deriv = trimSignal(deriv)

    grid = plt.GridSpec(2, 2, wspace=0.4, hspace=0.3)
    plt.subplot(grid[0,0:])
    plt.plot(x)
    plt.plot(peaks, x[peaks], 'x')

    plt.subplot(grid[1,0])
    plt.plot(deriv)

    plt.subplot(grid[1,1])
    plt.plot(trimmed_deriv)
    plt.show()
    
    return trimmed_deriv

deriv = verifyFit(data, rate)
```

```python
def bounceStats(time_deltas):
    g = 9.80665
    t = time_deltas / 2
    v = 1 / 2 * g * t
    # This works even though v is not constant.
    h = v * t  
    
    e = np.zeros(len(v) - 1)
    frac_loss = np.zeros(len(v) - 1)
    for i in range(len(v) - 1):
        e[i] = v[i+1] / v[i]
        frac_loss[i] = (h[i] - h[i+1]) / h[i]
    return v, h, e, frac_loss
```

```python
index = 2
v, h, e, loss = bounceStats(np.array(data[index]).astype(float))

plt.subplot(2,2,1:)
plt.plot(v, label='Velocity')
plt.plot(h, label='Height')
plt.legend()
plt.subplot(2,2,3)
plt.plot(e, label='Coef. of Restitution')
plt.legend()
plt.subplot(2,2,4)
plt.plot(frac_loss, label='Coef. of Restitution')
plt.legend()
plt.show()
```

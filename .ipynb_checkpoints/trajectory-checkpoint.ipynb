```python
#nbi:hide_in
from ipywidgets import interact, widgets, Label
import nbinteract as nbi
import numpy as np

class trajectory(object):
    def __init__(self):
        self.xpts = np.arange(0)
        self.ypts = np.arange(0) 
        
    def xvals(self, coef, m, angle, velocity):
        self.xpts = np.arange(0)
        self.ypts = np.arange(0) 
        dt = 0.1
        g = -9.8
        x = 0
        y = 0
        v = velocity
        v2 = v ** 2
        a = coef * v2 / m
        vx = v * np.cos(np.radians(angle))
        vy = v * np.sin(np.radians(angle))
        ax = a * np.cos(np.radians(angle)) * (-vx / np.abs(vx))
        ay = a * np.sin(np.radians(angle)) * (-vy / np.abs(vy)) + g

        while True:            
            angle = np.arctan(vy / (vx + 1e-8))
            v2 = vx ** 2 + vy ** 2
            a = coef * v2 / m
            ax = a * np.cos(angle) * (-vx / np.abs(vx + 1e-8))
            ay = a * np.sin(angle) * (-vy / np.abs(vy + 1e-8)) + g
            vx += ax * dt
            vy += ay * dt         
            x += vx * dt + 0.5 * ax * dt ** 2
            y += vy * dt + 0.5 * ay * dt ** 2

            if (y <= 0):
                break
            else:
                self.xpts = np.append(self.xpts, x)
                self.ypts = np.append(self.ypts, y)

        return self.xpts
    
    def yvals(self, foo):
        return self.ypts
    
    
a = trajectory()
coefficient = widgets.BoundedFloatText(value=0, min=0, step=0.0001, description=r'\(\frac{1}{2}C_dA\rho\)')
mass = widgets.BoundedFloatText(value=1, min=0.000001, description='mass [kg]')
xmax = 500
ymax = 500
opts = {'title': 'Projectile Trajectory',
        'xlim': (0,xmax),
        'ylim': (0,ymax),
        'xlabel': 'Horizontal Displacement [m]',
        'ylabel': 'Vertical Displacement [m]'}

nbi.line(a.xvals, a.yvals, coef=coefficient, m=mass, angle=(1,89), velocity=(1,100), options=opts)
```

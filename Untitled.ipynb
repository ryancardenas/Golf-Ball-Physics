```python
def showData(fn):
    rate, data = read(fn)
    data = data[:, np.newaxis]

    rms = np.sqrt(np.sum(data**2) / data.shape[0])
    rms = np.full(data.shape, rms)

    print('Sample Rate:', rate)
    plt.plot(data)
    plt.plot(2*rms)
    plt.plot(-2*rms)
    plt.show()
    
    return data, rate
    
fn = 'data\\raw_2_2.wav'
data, rate = showData(fn)
```

```python
def meanReduce(data, n=1):
    m = data.shape[0]
    dat = np.copy(data)

    for i in range(n):
        rms = np.sqrt(np.sum(dat**2) / m)

        dat = dat * (1 - rms / np.abs(dat + 1e-16))
    
    return dat

dat = meanReduce(data, n=10)
plt.plot(np.abs(dat))
plt.show()
```

```python
def absMeanReduce(data, n=100):
    m = data.shape[0]
    dat = np.abs(data)

    for i in range(n):
        rm = np.sum(np.abs(dat)) / m
    #     print('Iteration:', i+1, rm)

        dat = dat * (1 - rm / np.abs(dat + 1e-16))

    return dat
    
dat = absMeanReduce(data, n=100)
plt.plot(dat)
plt.show()
```

```python
def rmsCutoff(data, k=5):
    m = data.shape[0]
    dat = np.abs(data)

    rm = np.sum(np.abs(dat)) / m
    dat = np.where(dat < k*rm, 0, dat)
    
    return dat

dat = rmsCutoff(data, k=6)
plt.plot(dat)
plt.show()
```

```python
def trimSignal(signal):
    g = 9.80665
    h_drop = 0.815
    ti = signal[0]
    T_max = np.sqrt(8 * h_drop / g)
    print(T_max)
    started = False
    left = 0
    for i in range(len(signal)):
        tf = signal[i]

        # Set left trim once signal has started decreasing
        # and interval is less than the physical maximum time.
        if tf < ti and tf < T_max and started == False:
            started = True
            if ti < T_max:
                left = i - 1
            else:
                left = i
            
        # Set right trim if signal starts increasing.
        if tf > ti and started == True:
            return signal[left:i-1]
        
        ti = tf
        
    return signal[left:]
```

```python
def bounceStats(T):
    g = 9.80665
    v = 1 / 2 * g * T
    # This works even though v is not constant.
    h = 1 / 4 * v * T  
    
    e = np.zeros(len(v) - 1)
    frac_loss = np.zeros(len(v) - 1)
    for i in range(len(v) - 1):
        e[i] = v[i+1] / v[i]
        frac_loss[i] = (h[i] - h[i+1]) / h[i]
    return v, h, e, frac_loss
```

```python
v, h, e, loss = bounceStats(np.array(text[12][2:]).astype(float))
print(v, h)
```

```python
def plotStats(text):
    for index in range(len(text)):
        data = [row[2:] for row in text]
        if len(data) > 5:
            v, h, e, loss = bounceStats(np.array(data[index]).astype(float))
        else:
            continue
        
        print(text[index][0])
        plt.subplot(2,2,1)
        plt.plot(v, label='Velocity')
        plt.plot(h, label='Height')
        plt.legend()
        plt.subplot(2,2,3)
        plt.plot(e, label='Coef. of Restitution')
        plt.ylim(0.5, 1.0)
        plt.legend()
        plt.subplot(2,2,4)
        plt.plot(loss, label='Fraction of Energy Lost')
        plt.ylim(0, 0.5)
        plt.legend()
        plt.show()
        
plotStats(text)
```

```python
def aggregateStats(text):
    '''
    RETURNS:
    golfballs -- [name,[v, e, loss]] List of golf balls and velocity-dependent stats.
    '''
    names = sorted(set([x[0][0:-2] for x in text]))
    golfballs = [[name] for name in names]
    
    for i in range(len(names)):
        name = names[i]
        golfballs[i].append(np.array([[],[],[]]).T)
        
        for data in text:
            if name in data[0] and len(data) > 5:
                v, h, e, loss = bounceStats(np.array(data[1:]).astype(float))
                trial = [[x,y,z] for x,y,z in zip(v[:-1], e, loss)]
                golfballs[i][1] = np.concatenate((golfballs[i][1],trial))
        
        golfballs[i][1] = golfballs[i][1][golfballs[i][1][:,0].argsort()]
            
    return golfballs

golfballs = aggregateStats(text)
print(golfballs)
```

```python
def plotAggregateStats(text):
    golfballs = aggregateStats(text)
    for index in range(len(golfballs)):
        plt.suptitle('Ball: ' + golfballs[index][0], y=1.1, size=18)
        plt.subplot(2,1,1)
        x, y = golfballs[index][1][:,0], golfballs[index][1][:,1]
        z = np.polyfit(x, y, 1)
        r = z[0] * 20 + z[1]
        avg = 'Avg: ' + str(round(np.mean(y), 3)) +'\nStd: ' + str(round(np.std(y),3))
        plt.scatter(x, y, label=avg)
        plt.plot(x, np.poly1d(z)(x))
        plt.title('Est. Restitution at 20 [m/s]: '+str(round(r,3)))
        plt.xlabel('Velocity [m/s]')
        plt.ylabel('Coefficient of Restitution')
        plt.legend()

        plt.subplot(2,1,2)
        x, y = golfballs[index][1][:,0], golfballs[index][1][:,2]
        z = np.polyfit(x, y, 1)
        avg = 'Avg: ' + str(round(np.mean(y)*100, 1)) +'%'
        plt.scatter(x, y*100., color='orange',label=avg)
        plt.plot(x, np.poly1d(z*100)(x))
        plt.xlabel('Velocity [m/s]')
        plt.ylabel('Percent Energy Loss')
        plt.legend()
        plt.tight_layout()
        plt.savefig(golfballs[index][0]+'.png', bbox_inches = "tight")
        plt.show()
        
plotAggregateStats(text)
```

```python
def saveAggregateStats(text):
    golfballs = aggregateStats(text)
    vels = (1, 3, 5, 10, 20)
    rest = np.zeros((1,len(vels)))
    energy = np.zeros((1,len(vels)))
    for index in range(len(golfballs)):
        x, y = golfballs[index][1][:,0], golfballs[index][1][:,1]
        z = np.polyfit(x, y, 1)
        q = np.array([round(z[0] * v + z[1],3) for v in vels])
        rest = np.vstack((rest, q))
        
        x, y = golfballs[index][1][:,0], golfballs[index][1][:,2]
        z = np.polyfit(x, y, 1)
        r = np.array([round(z[0] * v + z[1],3) for v in vels])
        energy = np.vstack((energy, r))
        
    rest = np.delete(rest, 0, 0)
    energy = np.delete(energy, 0, 0)
    print(rest, '\n\n', energy)
    return rest, energy
        
rest, energy = saveAggregateStats(text)
```
